```python
# Calculate churn percentage
churn_percent = df['Churn'].value_counts(normalize=True) * 100

# Plot bar chart
ax = churn_percent.plot(kind='bar')

plt.title('Customer Churn Percentage')
plt.grid(True)
plt.xlabel('Churn')
plt.ylabel('Percentage')
plt.xticks(rotation=30)

# Add percentage labels on bars
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%',
               (p.get_x() + p.get_width()/2.,p.get_height()),
               ha='center',va='bottom')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Churn value
churn_value = 26  # %

fig, ax = plt.subplots(figsize=(8, 4))

# Create angle range
angles = np.linspace(0, np.pi, 100)

# Draw gauge background
ax.plot(angles, [1]*100, linewidth=30, color='lightgrey')

# Color zones
ax.plot(np.linspace(0, np.pi*0.10, 30), [1]*30, linewidth=30, color='green')   # <10%
ax.plot(np.linspace(np.pi*0.10, np.pi*0.20, 30), [1]*30, linewidth=30, color='yellow') # 10–20%
ax.plot(np.linspace(np.pi*0.20, np.pi, 40), [1]*40, linewidth=30, color='red') # >25%

# Needle
needle_angle = np.pi * churn_value / 100
ax.plot([needle_angle, needle_angle], [0, 1], color='black', linewidth=3)

# Text
ax.text(0, -0.2, f"Churn Rate: {churn_value}%", ha='center', fontsize=14, fontweight='bold')
ax.text(0, -0.35, "Status: HIGH RISK 🚨", ha='center', fontsize=12, color='red')

# Clean up
ax.set_theta_zero_location('W')
ax.set_theta_direction(-1)
ax.set_axis_off()

plt.title("Customer Churn Risk Gauge", fontsize=16)
plt.show()
```

```python
#2.CHURN BY CONTRACT TYPE(Does contract type affect churn?)
```

```python
df.groupby('Contract')['Churn'].value_counts()
```

```python
contract_churn_pct = (
    df.groupby('Contract')['Churn']
    .value_counts(normalize=True)
    .rename('Percentage')
    .mul(100)
    .reset_index()
)

contract_churn_pct
```

```python
churn_yes = contract_churn_pct[contract_churn_pct['Churn'] == 'Yes']
churn_yes
```

```python
churn_yes_sorted = churn_yes.sort_values(by='Percentage',ascending=False)
churn_yes_sorted
```

```python
ax = churn_yes_sorted.plot(
    kind='bar',
    x ='Contract',
    y ='Percentage',
    legend =False
)

plt.title('Churn Percentage by Contract type')
plt.xlabel('Contaract type')
plt.ylabel('Churn Percentage')
plt.xticks(rotation=30)

# Add Percentage labels
for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%',
             (p.get_x() + p.get_width()/2,p.get_height()),
             ha='center',va='bottom')
plt.show()
```

```python
#3.Does customer age (tenure) affect churn?
df['Tenure_Group'] = pd.cut(
    df['tenure'],
    bins=[0,12,24,48,60,72],
    labels=['0-1 Year','1-2 Years','2-4 Years','4-5 Years','5+ Years']
)
```

```python
tenure_churn = (
    df.groupby('Tenure_Group')['Churn']
    .value_counts(normalize=True)
    .mul(100)
    .rename('Percentage')
    .reset_index()
)
```

```python
tenure_churn_yes = tenure_churn[tenure_churn['Churn'] == 'Yes']
tenure_churn_yes
```

```python
tenure_churn_sorted = tenure_churn_yes.sort_values(by='Percentage',
                                                    ascending=False
                                                  )
tenure_churn_sorted
```

```python
ax = tenure_churn_sorted.plot(
    kind='bar',
    x='Tenure_Group',
    y='Percentage',
    legend=False
)

plt.title('Churn Percentage by tenure Group')
plt.xlabel('Tenure Group')
plt.ylabel('Churn Percentage')
plt.xticks(rotation=30)

for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%',
               (p.get_x() + p.get_width()/2,p.get_height()),
               ha='center',va='bottom')
plt.show()
```
